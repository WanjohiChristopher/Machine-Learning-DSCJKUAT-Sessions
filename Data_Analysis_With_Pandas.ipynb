## Core components of pandas: Series and DataFrames

The primary two components of pandas are : the Series and DataFrame.

A Series is essentially a column, 

A DataFrame is a multi-dimensional table made up of a collection of Series.

```python
import pandas as pd
#creating a dictionary{key:values}
data = {
    'apples': [3, 2, 0, 1], 
    'oranges': [0, 3, 7, 2],
    "Mangoes":[6,7,10,3]
}
```

```python
#pass the data variable to the  pandas DataFrame constructor
df=pd.DataFrame(data)
#viewing the data
df.head()
```

```python
# create index column by yourself
purchases = pd.DataFrame(data, index=['James', 'Robert', 'Lily', 'Mary'])

purchases
```

```python
#we can use loc in pandas to acces individual person details
purchases.loc['Robert']
```

```python
# we can also set index to be apples for example
df = purchases.set_index('apples')

df
```

```python
#drop values from a row(axis=0)
df.drop([3,1],axis=0,inplace=True)

df

```

```python
#dropping columns axis=1
df.drop('oranges', axis=1,inplace=True)
df
```

```python
df=df.rank()
df
```

```python
puchases=purchases.sort_values(by='oranges')
purchases
```

## Pivot Tables
    A pivot table is a similar operation that is commonly seen in spreadsheets and other programs that operate on tabular data. The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

    A Pivot Table is used to summarise, sort, reorganise, group, count, total or average data stored in a table. It allows us to transform columns into rows and rows into columns. It allows grouping by any field (column), and using advanced calculations on them.

    pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False)[source]¶ Create a spreadsheet-style pivot table as a DataFrame.

    The levels in the pivot table will be stored in MultiIndex objects (hierarchical indexes) on the index and columns of the result DataFrame.



```python
# Applying multiple aggregate functions to a group

pd.pivot_table(shopify, index=['Item_Type', 'Item_Fat_Content'], aggfunc={'Item_Outlet_Sales': sum}).head(20)
```

### Data Cleaning /Data Wrangling

# Removing null values

Data Scientists and Analysts regularly face the dilemma of dropping or imputing null values, and is a decision that requires intimate knowledge of your data and its context. Overall, removing null data is only suggested if you have a small amount of missing data.

```python
import missingno as msno

# Gives a bar chart of the missing values
msno.bar(shopify)
```

```python
# msno.heatmap(shopify, cmap='YlGnBu')
```

```python
#dropping misssing values
# shopify.dropna()
```

```python
#filling missing values with either mean,mode,median
shopify['Item_Weight'].fillna(shopify['Item_Weight'].mean(),inplace=True)
```

```python
shopify['Outlet_Size'].fillna(shopify['Outlet_Size'].mode()[0],inplace=True)
```

```python
shopify['Item_Visibility'].fillna(shopify['Item_Visibility'].median(),inplace=True)
```

```python
#using foward filling
shopify['Item_MRP'].fillna(method='ffill', inplace=True)
```

```python
# using backward filling
# shopify['Item_MRP '].fillna(method='bfill', inplace=True)
```
